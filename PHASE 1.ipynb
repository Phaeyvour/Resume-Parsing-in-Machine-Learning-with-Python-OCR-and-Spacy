```python
from pathlib import Path
from PIL import Image
import pytesseract
import sys
pytesseract.pytesseract.tesseract_cmd = (r"C:\Program Files\Tesseract-OCR\tesseract.exe")

def ocr_image(img_path: str | Path, lang: str = "eng") -> str:
    img_path = Path(img_path)
    if not img_path.is_file():
        raise FileNotFoundError(f"Image not found: {img_path.resolve()}")

    try:
        with Image.open(img_path) as img:
            gray = img.convert("L")            # quick pre‑processing boost
            return pytesseract.image_to_string(gray, lang=lang)
    except Exception as e:
        print(f"[OCR] Error processing {img_path.name}: {e}", file=sys.stderr)
        raise

if __name__ == "__main__":
    text = ocr_image(r"C:\Users\HP\Downloads\GemtEiYXQAAo8bg.jpg")
    print(text)
```
